```python
novo_preço_07 = df_produto['preco'].apply(lambda x: x+x*0.1, convert_dtype = False)
novo_preço_14 = df_produto['preco'].apply(lambda x: x+x*0.15, convert_dtype = False)
novo_preço_21 = df_produto['preco'].apply(lambda x: x+x*0.2, convert_dtype = False)
novo_preço_28 = df_produto['preco'].apply(lambda x: x+x*0.25, convert_dtype = False)

novo_preço_07 = novo_preço_07.to_frame()
novo_preço_14 = novo_preço_14.to_frame()
novo_preço_21 = novo_preço_21.to_frame()
novo_preço_28 = novo_preço_28.to_frame()
```
